```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
A = pd.read_csv("C:/Users/Pawan/Downloads/50_Startups.csv")
```

# Define Y and X

```python
Y = A[["PROFIT"]]
X = A.drop(labels=["PROFIT"],axis=1)
```

# Preprocessing

```python
from PM8_weekend import preprocessing,replacer
replacer(A)
Xnew = preprocessing(X)
```

```python
Xnew
```

# Divide Data in training and testing set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=23)
```

# Create a NN

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(hidden_layer_sizes=(100,100,100),random_state=21)
nn = mlpr.fit(xtrain,ytrain)
pred_ts = nn.predict(xtest)
pred_tr = nn.predict(xtrain)

from sklearn.metrics import mean_squared_error
ts_err = mean_squared_error(ytest,pred_ts)
tr_err = mean_squared_error(ytrain,pred_tr)
```
